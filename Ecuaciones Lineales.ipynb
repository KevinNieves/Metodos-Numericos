```python
import math 
import numpy as np
import pylab as pl
from pylab import *

class MyClass(object):
    
    def  funcion1(x):
        Fx=((-3.0*x)/2.0) + 9.0
        return Fx
    funcion1 = staticmethod(funcion1)
    
    def  funcion2(x):
        Fx=(x/2.0) + 1.0
        return Fx
    funcion2 = staticmethod(funcion2)
    
    def  funcion3(x):
        Fx= 1 + (x/2)
        return Fx
    funcion3 = staticmethod(funcion3)
    
    def  funcion4(x):
        Fx= 0.5 +(x/2)
        return Fx
    funcion4 = staticmethod(funcion4)
    
    def  funcion5(x):
        Fx= 1 + (x/2)
        return Fx
    funcion5 = staticmethod(funcion5)
    
    def  funcion6(x):
        Fx= (2+x)/2
        return Fx
    funcion6 = staticmethod(funcion6)
    
    def  funcion7(x):
        Fx= (1.1 + (2.3*x)/5)
        return Fx
    funcion7 = staticmethod(funcion7)
    
    def  funcion8(x):
        Fx= 1 + x/2
        return Fx
    funcion8 = staticmethod(funcion8)

    def graficar():
        i=0.0
        fx1=[]
        fx2=[]
        fx3=[]
        fx4=[]
        fx5=[]
        fx6=[]
        fx7=[]
        fx8=[]
        iteracion=[]
        while i<=5.0:
            
            fx1.append(MyClass.funcion1(i))
            fx2.append(MyClass.funcion2(i))
            
            fx3.append(MyClass.funcion3(i))
            fx4.append(MyClass.funcion4(i))
            
            fx5.append(MyClass.funcion5(i))
            fx6.append(MyClass.funcion6(i))
            
            fx7.append(MyClass.funcion7(i))
            fx8.append(MyClass.funcion8(i))
            
            iteracion.append(i)
            i=i+0.5

        plt.subplot(2,1,1)
        
        plt.plot(iteracion,fx1,label = "Funcion1")
        plt.plot(iteracion,fx2,label = "funcion2")
        plt.title("Funciones:([-3.0X/2.0] + 9.0)  &  ([X/2.0] + 1.0)")
        plt.grid(True)
        
        plt.subplot(2,1,2)
        
        plt.plot(fx3,label = "Funcion3")
        plt.plot(fx4,label = "funcion4")
        plt.title("representacion cuando el sistema no tiene solucion")
        plt.grid(True)
        
        plt.show()
        
        plt.subplot(2,1,1)
        
        plt.plot(fx5,label = "Funcion5")
        plt.plot(fx6,label = "funcion6")
        plt.title("cuando el sistema tiene infinitas soluciones")
        plt.grid(True)
        
        plt.subplot(2,1,2)
        
        plt.plot(fx7,label = "Funcion7")
        plt.plot(fx8,label = "funcion8")
        plt.title("sistema mal condicionado, pendientes muy cercanas ")
        plt.grid(True)
       
        plt.show()
        
    graficar = staticmethod(graficar)
    
j=MyClass.graficar()
```
